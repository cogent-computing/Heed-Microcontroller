```python
#Initialise
import random
import pandas as pd

print("----------Initialising Data-Set-----------")
test_env = ControllEnvironment(day_nr=18, actual_hour=0, battery_power=21.1, pv_scale=1.0,
                               priorities=[1, 2, 3, 4, 5, 6, 7])
print("Starting Scenario")
display(test_env.get_starting_scenario())

# Random Control
control_dict = {"Nursery1_Lights_Quota": [round(random.random(), 2) for x in range(0, 24)],
                "Nursery1_Sockets_Quota": [round(random.random(), 2) for x in range(0, 24)],
                "Nursery2_Lights_Quota": [round(random.random(), 2) for x in range(0, 24)],
                "Nursery2_Sockets_Quota": [round(random.random(), 2) for x in range(0, 24)],
                "Playground_Lights_Quota": [round(random.random(), 2) for x in range(0, 24)],
                "Playground_Sockets_Quota": [round(random.random(), 2) for x in range(0, 24)],
                "Streetlights_Quota": [round(random.random(), 2) for x in range(0, 24)]
                }
print("Control Dictionary")
display(pd.DataFrame.from_dict(control_dict))

output_scenario = test_env.step_24h(control_dict, battery_power=21.1, pv_scale=1.0, step_type="percentage")

print("Results of Control Scenario 1")
display(output_scenario)

# All Off
control_dict2 = {"Nursery1_Lights_Quota": [0 for x in range(0, 24)],
                 "Nursery1_Sockets_Quota": [0 for x in range(0, 24)],
                 "Nursery2_Lights_Quota": [0 for x in range(0, 24)],
                 "Nursery2_Sockets_Quota": [0 for x in range(0, 24)],
                 "Playground_Lights_Quota": [0 for x in range(0, 24)],
                 "Playground_Sockets_Quota": [0 for x in range(0, 24)],
                 "Streetlights_Quota": [0 for x in range(0, 24)]
                 }

print("Control Dictionary")
display(pd.DataFrame.from_dict(control_dict2))

output_scenario2 = test_env.step_24h(control_dict2, battery_power=21.1, pv_scale=1.0, step_type="percentage")

print("Results of Control Scenario 2")
display(output_scenario2)

#Save results
#test_env.save_result(output_scenario,"./out.csv")
```
